```python
#Resize gambar ke 300x400
def resize(img):
  w = 300
  h = 400
  resize_img = cv2.resize(img, (w,h))
  return resize_img

resized_img = resize(img)
cv2_imshow(resized_img)
```

```python
# Crop 200x250 di bagian tengah
def crop_center_resize(img, width, height):
    h, w = img.shape[:2]
    center_x, center_y = w // 2, h // 2

    # Ambil kotak tengah
    crop_start_x = max(0, center_x - width // 2)
    crop_start_y = max(0, center_y - height // 2)
    crop_end_x = min(w, center_x + width // 2)
    crop_end_y = min(h, center_y + height // 2)

    cropped_img = img[crop_start_y:crop_end_y, crop_start_x:crop_end_x]

    # Resize hasil crop
    resized_img = cv2.resize(cropped_img, (width, height))
    return resized_img

img_path = 'Bulet.png'

# Baca gambar
img = cv2.imread(img_path)

# Pastikan gambar berhasil dimuat
if img is not None:
    resized_img = resize(img)
    crop_img = crop_center_resize(resized_img, 200, 250)

    # Tampilkan gambar hasil crop
    from google.colab.patches import cv2_imshow
    cv2_imshow(crop_img)
else:
    print("Gambar tidak ditemukan. Periksa kembali path ke file gambar.")
```

```python
# Buat BnW
def convert_to_grayscale(img):

    grayscale_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    return grayscale_img

grayscale_image = convert_to_grayscale(crop_img)
cv2_imshow(grayscale_image)
```

```python
def preprocessing(img_path):
  #Import Gambar
    img = cv2.imread(img_path)

  #resize
    w = 300
    h = 400
    img = cv2.resize(img, (w,h))

  #crop
    width = 200
    height = 250
    h, w = img.shape[:2]

    # Cari center
    center_x, center_y = w // 2, h // 2

    # Ambil kotak tengah
    crop_start_x, crop_start_y = max(0, center_x - width // 2), max(0, center_y - height // 2)
    crop_end_x, crop_end_y = min(w, center_x + width // 2), min(h, center_y + height // 2)

    cropped_img = img[crop_start_y:crop_end_y, crop_start_x:crop_end_x]

    # Resize
    img = cv2.resize(cropped_img, (width, height))

    #BnW
    grayscale_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    return grayscale_img
```

```python
tes = preprocessing("Bulet.png")
cv2_imshow(tes)
```
